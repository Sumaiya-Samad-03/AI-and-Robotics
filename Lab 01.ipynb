```python
Student = {
    'S1':{'Rollno':'CND-02529212', 'Name':'Sumaiya', "CNIC":'37405-12345678-9'},
    'S2':{'Rollno':'CND-02529213', 'Name':'Sana', 'CNIC':'37405-98765432-1'},
    }

def student_detail(CNIC):
  for student in Student.values():
    if student['CNIC'] == CNIC:
      return student
  #if CNIC in Student:
   # return Student[CNIC]
  else:
    return "CNIC not found"

CNIC = str(input("Enter your CNIC number (xxxxx-xxxxxxxx-x): "))
print(CNIC)
print(student_detail(CNIC))
```

```python
n = int(input("Enter a number: "))
for i in range(0,11):
  print(f"{n} x {i} = {n*i}")
```

```python
n = int(input("Enter a number: "))
limit = int(input("Enter the upper limit for the table: "))
for i in range(0, limit + 1):
  print(f"{n} x {i} = {n*i}")
```
