###Inheritance
 - Single Inheritance
 - Multiple Inheritance - 1 child but multiple parents
 - Multilevel - step-by-step parent class, child class, grandchild class
 - Hierarchical Inheritance - one parent but multiple child classes
 - Hibrid Inheritance - one program has multiple inheritance at the same time having multiple parents and multiple children

# Single Inheritance

```python
class Animal:
  def sound(self):
    print("Animals makes sound")
class Dog(Animal):
  def barking(self):
    print("Ghwab ghwab")
d = Dog()
d.sound()
d.barking()
b = Animal()
b.sound()
```
