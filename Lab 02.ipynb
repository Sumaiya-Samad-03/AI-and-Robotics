Exception Handling


##what is exception handling?
Exception handling remove the chances of error and prevent the code crashing. This adds meaning to the error; like what type of error is appearing.

```python
a = 3
b = 0
c = a/b
```

```python
try:
  a = 4
  b = 0
  c = a/b
except ZeroDivisionError:
 print("Enter a valid value")
```

```python
try:
  a = 2
  b = 0
  c = a/b
except Exception as e:
  print(e)
```

```python
try:
  Num1 = int(input("Enter a number"))
  Num2 = int(input("Enter a number"))
  Result = Num1/Num2
  print(Result)
except ZeroDivisionError:
  print("Enter a number other than zero")
except ValueError:
  print("Enter a number!")
```

#OOPs
the way we will use the classes, method,
It has four pillars
1. Abstraction - No idea what is happening in the background
2. Encapsulation - Hide the information/Privacy
3. Polymorphism - Different name but same functions
4. Inheritance - Parents inherit data to the child classes

```python
name = "Sumaiya"
age = 22
print("My name is ",name,"and I am ",age, "years old." )
```

```python
print(f"My name is {name} and I am {age} years old.")
```

```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def detail(self):
        print("The student name is ", self.name, "and is ", self.age, "years old." )
        print(f"The student name is {self.name} and is {self.age} years old")
S1 = Student("Sumaiya", 22)
S2 = Student("Ayesha", 19)
S1.detail()
S2.detail()
```

```python
class Employee:
  def __init__(self, name, qualification):
    self.name = name
    self.qualification = qualification
  def detail(self):
    print(f"The employee name is {self.name} and is {self.qualification} qualified.")

E1 = Employee(input("Enter your name: "), input("Enter your qualification: "))
E1.detail()
```

- Encapsulation

```python
class Bankacc:
  def __init__(self, name, acc_no, balance):
    self.name = name
    self.__balance = balance
    self.acc_no = acc_no
  def Show_Bankdetails(self):
    print("Your balance is ", self.__balance)

obj = Bankacc("Sumaiya", 123456789, 10000)
print(obj.name)
print(obj.acc_no)

try:
  print(obj.__balance)
except AttributeError:
  print("You can't access balance of any person.")
obj.Show_Bankdetails()
```
