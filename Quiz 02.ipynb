```python
#Compute values of 00 and 01
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv('dataset.csv')
x = df['x'].values
y = df['y'].values

# 1. Initialize parameters
w = 0.0  # weight (theta1)
b = 0.0  # bias (theta0)
learning_rate = 0.01
iterations = 1000

# 2. Gradient Descent
for i in range(iterations):
    # Predictions: y = w*x + b
    y_pred = w * x + b

    # Gradients
    dw = (2/len(x)) * np.sum(x * (y_pred - y))
    db = (2/len(x)) * np.sum(y_pred - y)

    # Update parameters
    w = w - learning_rate * dw
    b = b - learning_rate * db

print(f"Learned: w(Θ₁) = {w:.4f}, b(Θ₀) = {b:.4f}")

# 3. Final predictions
y_final_pred = w * x + b

# 4. Plot using matplotlib (simplest)
plt.figure(figsize=(10,6))
plt.scatter(x, y, alpha=0.7, label='Data', s=50)
plt.plot(x, y_final_pred, 'r-', linewidth=3, label=f'y = {w:.2f}x + {b:.2f}')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Univariate Linear Regression (Gradient Descent)')
#plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# 5. Seaborn version (if you want)
plt.figure(figsize=(10,6))
sns.scatterplot(x=x, y=y, alpha=0.7, s=50)
plt.plot(x, y_final_pred, 'r-', linewidth=3, label=f'GD: y={w:.2f}x+{b:.2f}')
plt.legend()
plt.title('Gradient Descent Linear Regression')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
#using matrix notation, derive expression for the closed-form solution of linear regression and then implement
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('dataset.csv')
X = np.c_[np.ones(len(df)), df['x']]  # Matrix with bias
theta = np.linalg.inv(X.T@X)@X.T@y   # Closed-form θ

plt.scatter(df['x'], df['y']); plt.plot(df['x'], X@theta, 'r-', lw=3)
plt.title(f'y = {theta[1]:.2f}x + {theta[0]:.2f}'); plt.show()
print(f"θ₀={theta[0]:.3f}, θ₁={theta[1]:.3f}")
```
