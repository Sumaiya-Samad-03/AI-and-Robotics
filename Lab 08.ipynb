##Selection Sort
a simple comparison based sorting algorithm. It finds the smallest value and then start sorting from there instead moving in pairs. This sorting method is faster then the bubble sort.


```python
alphabets = ['h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'a', 'b', 'c', 'd', 'e', 'f', 'g', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']
def selection(alphabets):
  n = len(alphabets)
  for i in range(n):
    index = i
    for j in range(i+1, n):
      if alphabets[j] < alphabets[index]:
        index = j
    alphabets[i], alphabets[index] = alphabets[index], alphabets[i]
  return alphabets
sa = selection(alphabets)
print(sa)
```

##Recurrsion


```python
def countdown(n):
  if n == 0:
    print("Stopped the countdown.")
  else:
    print(n)
    countdown(n-1)
countdown(10)
```

```python
def factorial(n):
  if n==0:
    return 1
  else:
    return n * factorial(n-1)
print(factorial(7))
```

##Fibonacci Series
sequence of numbers where each number is the sum of two preceding ones
###Formula
fib[n]

```python
def fib(n):
  if n == 0:
    return 0
  elif n == 1:
    return 1
  else:
    return fib(n-1) + fib(n-2)
print(fib(10))
```
